```python
# Importing required libraries and evaluation functions
import sys
sys.path.append("../src")

import pandas as pd
from monitoring import simulate_shift, calculate_psi

from models import train_xgb_model
```

```python
# Loading the feature-engineered dataset
df = pd.read_csv("../data/featured_data.csv")

# Training the XGBoost model and getting test data along with prediction probabilities
model, X_test, y_test, y_prob = train_xgb_model(df)

# Simulating distribution shift on the test data
# This helps us understand how the model behaves when data changes over time
X_shifted = simulate_shift(X_test)

# Generating predictions on the shifted data
# We use probabilities to compare model behavior before and after shift
y_prob_shifted = model.predict_proba(X_shifted)[:,1]

# Calculating PSI (Population Stability Index)
# PSI helps measure how much the data distribution has changed
psi = calculate_psi(y_prob, y_prob_shifted)

# Printing PSI value to check if there is significant drift in the data
print("PSI:", psi)
```
