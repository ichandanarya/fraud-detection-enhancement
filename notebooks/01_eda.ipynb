## Importing required libraries for data analysis and visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset from the data folder
df = pd.read_csv("../data/creditcard.csv")

# Checking dataset size (rows, columns)
print(df.shape)

# Displaying first few rows to understand structure
print(df.head())


# Visualizing how many fraud vs non-fraud transactions are present
# This helps to understand class imbalance (very important in fraud detection)
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Non-Fraud")
plt.show()

# Printing percentage distribution of classes
# This confirms how imbalanced the dataset is
print(df['Class'].value_counts(normalize=True))

# Visualizing distribution of transaction amounts
# Helps understand if there are extreme values or patterns
sns.histplot(df['Amount'], bins=50)
plt.title("Transaction Amount")
plt.show()
```
